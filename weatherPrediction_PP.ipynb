```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=.1)
```

```python
def create_predictions(predictors,core_weather, reg, target, t_day,N_to_Z):
    core_weather["target"]=core_weather.shift(-t_day)[target]
    core_weather=core_weather.iloc[:-t_day,:].copy()

    
    train= core_weather[:"2015-12-31"]
    test= core_weather["2016-1-1":]
    #train= core_weather[:"2018-12-31"]
    #test= core_weather["2019-1-1":]
    reg.fit(train[predictors],train["target"])
    predictions= reg.predict(test[predictors])
    combined= pd.concat([test["target"],pd.Series(predictions, index=test.index)],axis=1)
    combined.columns = ["actual","predictions"]
    
    from sklearn.metrics import mean_absolute_error
    
    if N_to_Z:
        combined["predictions"].where(combined["predictions"] > 0, 0, inplace=True)
        error= mean_absolute_error(combined["actual"],combined["predictions"])
    else:
        error= mean_absolute_error(test["target"],predictions)

    error_p= mean_absolute_error(test[target],test["target"])
    return error,error_p, combined
```

```python
r_day=105
my_target="temp_mean"
negative_to_zero=False
#False  do not change
#True change
for x in range(1,r_day+1):
    if x==1:
        error,error_p, combined= create_predictions(predictors,core_weather.copy(),reg, my_target ,x, negative_to_zero)
        new_combined=combined[["actual"]].copy()
        new_combined["day"+"_"+str(x)]=combined[["predictions"]].copy()
    else:
        error,error_p, combined= create_predictions(predictors,core_weather.copy(),reg, my_target ,x, negative_to_zero)
        new_combined["day"+"_"+str(x)]=combined[["predictions"]].copy()
    
    if x == r_day and r_day != 1:
        new_combined=new_combined.iloc[:-(r_day-1),:].copy()
        
    

    print("for r_day",x,"error:",error,"error_p:",error_p,"error/error_p",round((1-error/error_p),3))

import datetime
#new_combined.iloc[0,2]
#new_combined.iat[2,0]
#new_combined.index[0]+ datetime.timedelta(days=1)
new_combined
my_day=new_combined.index[0]
my_counter=1
new_combined["real_predictions"]=0
for y in new_combined.index:
    #print(my_day+datetime.timedelta(days=r_day) ==y)
    #print(my_day+datetime.timedelta(days=r_day) ,y)
    if my_day+datetime.timedelta(days=r_day) ==y:
        my_day = y

    new_combined.at[y,"real_predictions"]=new_combined.at[my_day,"day_"+str(my_counter)]
    my_counter=my_counter+1
    if my_counter== (r_day+1):
        my_counter=1


new_combined=new_combined.shift(1)
new_combined=new_combined.iloc[1:,:].copy()
```
